```python
import os
import requests
from bs4 import BeautifulSoup
import json
import re
# utf-8
urls = ["https://yu-rin.com/talent/mens-a.html", "https://yu-rin.com/talent/mens-ka.html", "https://yu-rin.com/talent/mens-sa.html", "https://yu-rin.com/talent/mens-ta.html", "https://yu-rin.com/talent/mens-na.html", "https://yu-rin.com/talent/mens-ha.html", "https://yu-rin.com/talent/mens-ma.html", "https://yu-rin.com/talent/mens-ya.html", "https://yu-rin.com/talent/mens-ra.html", "https://yu-rin.com/talent/mens-wa.html", "https://yu-rin.com/talent/azukari-mens.html", "https://yu-rin.com/talent/womens-a.html", "https://yu-rin.com/talent/womens-ka.html", "https://yu-rin.com/talent/womens-sa.html", "https://yu-rin.com/talent/womens-ta.html", "https://yu-rin.com/talent/womens-na.html", "https://yu-rin.com/talent/womens-ha.html", "https://yu-rin.com/talent/womens-ma.html", "https://yu-rin.com/talent/womens-ya.html", "https://yu-rin.com/talent/womens-ra.html", "https://yu-rin.com/talent/womens-wa.html", "https://yu-rin.com/talent/azukari-womens.html"]
actors = []
for url in urls:
    response = requests.get(url)# html lang="ja",
    content = response.content.decode("utf-8")
    soup = BeautifulSoup(content, "html.parser")
    talents = soup.find_all("div", class_="md-talent")
    for talent in talents:
        actor = {}       
        prefix = "https://yu-rin.com/talent/"
        a = talent.find("a")
        actor["name"] = a.text.replace("\u3000"," ").strip()
        actor["homepage"] = a["href"].replace("./", prefix)
        actor["sex"] = "Female" if "womens" in url else "Male"
        actors.append(actor)
actors
```

```python
import re
import json
def get_voices(homepage:str):
    response = requests.get(homepage)
    content = response.content
    soup = BeautifulSoup(content, "html.parser")
    page_data = json.loads(content)
    voices = []
    voice_div = soup.find("div", class_ = "voice-list")
    voice_samples = voice_div.find_all("a")
    for voice_sample in voice_samples:
        voice = {}
        voice["url"] = voice_sample["href"].replace("./", "https://yu-rin.com/talent/")
        voice["name"] = voice_sample["src"].split("/")[-1]
        voice["description"] = voice_sample.text.strip()
        voices.append(voice)
    return voices

get_voices("https://yu-rin.com/talent/yourin_my0427.html")
#看起来得试图下载js
# import tqdm
# pbar = tqdm.tqdm(actors)
# for actor in pbar:
#     try:
#         actor["voices"] = get_voices(actor["homepage"])
#     except:
#         actor["voices"] = []
#     pbar.set_description(str(len(actor["voices"])))
# actors
```
